```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#sourse from : https://www.kaggle.com/code/mrappplg/ads-2-amazon-products-sales-dataset-2023/input?select=All+Electronics.csv

#import dataset
df = pd.read_csv('/Users/ccguaa/Data analyst/Dataset/Amazon All electronics sale /All Electronics.csv')
df
```

```python
#Take a look at overall dataframe
df.describe()
```

```python
#Checking for datatype in each column

df.info()

#Every columns are object(or string) type, so the first step is convert each column to correct format
```

```python
#Search for str value, then convert to 0
search_value = 'Get'
result = df[df['ratings'].isin([search_value])]
df.loc[df['ratings'] == 'Get', 'ratings'] = 0

# Convert the 'ratings' column to numeric type
df['ratings'] = pd.to_numeric(df['ratings'])



df['ratings']



```

```python
# Dictionary of values to search for and their replacements in 'no_of_ratings' column
replace_values = {'315.01': '315', '4,499.01': '4,499', '4,999.01': '4,999'}

# Replace specific values in the 'no_of_ratings' column
df['no_of_ratings'].replace(replace_values, inplace=True)

# Convert the 'no_of_ratings' column to numeric type
df['no_of_ratings'] = pd.to_numeric(df['no_of_ratings'].replace(',', ''), errors='coerce')
df['no_of_ratings'] = df['no_of_ratings'].fillna(0).astype(int)


# search for value 'FREE Delivery by Amazon' and replace with 0 value
search_value = 'FREE Delivery by Amazon'
result = df[df['no_of_ratings'].isin([search_value])]
df.loc[df['no_of_ratings'] == 'FREE Delivery by Amazon', 'no_of_ratings'] = 0


df['no_of_ratings']
```

```python

# Remove the '₹', ',' signs from the 'Price' column
df['discount_price'] = df['discount_price'].replace('[₹,]', '', regex=True)

df['discount_price'] = df['discount_price'].fillna(0).astype(float)

df['discount_price']
```

```python
# Remove the '₹', ',' signs from the 'Price' column
df['actual_price'] = df['actual_price'].replace('[₹,]', '', regex=True)


# Convert the 'no_of_ratings' column to numeric type
df['actual_price'] = df['actual_price'].fillna(0).astype(float)
df['actual_price'] = df['actual_price'].round(0)


df['actual_price']
```

```python
df.info()
```

```python
#Find percentage discount and create new column 'percentage_discount', then round up to 2 digits
position = df.columns.get_loc('actual_price') + 1
df.insert(position, 'percentage_discount', ((df['actual_price'] - df['discount_price']) / df['actual_price']) * 100)


df['percentage_discount']
```

```python
# Specify the columns you want to check for non-numeric values
column_to_check = 'ratings'

# Convert the column to numeric, coercing errors to NaN
df[column_to_check] = pd.to_numeric(df[column_to_check], errors='coerce')

# Remove rows where the specified column contains non-numeric values
df = df.dropna(subset=[column_to_check])

# Remove rows where the specified column contains zero
df = df[df[column_to_check] != 0]


df['ratings']
```

```python

# Specify the column you want to check for non-numeric values
column_to_check = 'no_of_ratings'

# Convert the column to numeric, coercing errors to NaN, and remove rows
df = df[pd.to_numeric(df[column_to_check], errors='coerce').notna()]
df = df[df[column_to_check] != 0]

df['no_of_ratings']
```

```python
# Specify the column you want to check for non-numeric values
column_to_check = 'actual_price'

# Remove rows where the specified column contains non-numeric values or is equal to 0
df = df[pd.to_numeric(df[column_to_check], errors='coerce').notna() & (df[column_to_check] != 0)]

# Display the 'actual_price' column from the resulting DataFrame
df['actual_price']
```

```python
#Select needed columns for the analysis
selected_columns = df[['name', 'ratings', 'no_of_ratings', 'discount_price', 'actual_price']].reset_index(drop=True)
selected_columns
```

```python
#selected_columns = selected_columns.drop('percentage_discount', axis=1)
selected_columns
```

```python
#reset index to start from row 1

selected_columns.index = selected_columns.index + 1
selected_columns
```

```python



# Calculate the percentage discount
selected_columns['percentage_discount'] = (((selected_columns['actual_price'] - selected_columns['discount_price']) / selected_columns['actual_price']) * 100).round(2)

# Convert 'ratings' to numeric type
selected_columns['ratings'] = pd.to_numeric(selected_columns['ratings'], errors='coerce')

# Convert 'actual_price' and 'discount_price' to numeric type
selected_columns['actual_price'] = pd.to_numeric(selected_columns['actual_price'], errors='coerce')
selected_columns['discount_price'] = pd.to_numeric(selected_columns['discount_price'], errors='coerce')

# Calculate total sale amount
selected_columns['total_sale'] = selected_columns['no_of_ratings'] * selected_columns['actual_price']

# Due to when 'discount_price' column value is 0 the 'percentage_discount' column value will be 100% 
# So I replace values 100% in the 'percentage_discount' column to 0
selected_columns.loc[selected_columns['percentage_discount'] == 100, 'percentage_discount'] = 0



```

```python
selected_columns['percentage_discount']
```

```python


# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot (x ='percentage_discount', y='total_sale', hue='ratings', size='ratings', data=selected_columns, palette='viridis')

# Set labels and title
plt.xlabel('Percentage Discount (%)')
plt.ylabel('Total Sale Amount (Indian rupee)')
plt.title('Effect of Discount Percentage on Total Sale Amount Based on Ratings')

# Show the plot
plt.show()
```
